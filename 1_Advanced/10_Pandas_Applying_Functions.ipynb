# .apply() - Calculate Project Salary Next Year

```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

# use .apply()

```python
df_salary = df[pd.notna(df['salary_year_avg'])]

def projected_salary(salary):
    return salary * 1.03

df['salary_year_inflated'] = df_salary['salary_year_avg'].apply(projected_salary)

df_salary[['salary_year_avg', 'salary_year_inflated']]
```

# use lambda

```python
df_salary['salary_year_inflated'] = df_salary['salary_year_avg'].apply(lambda salary: salary*1.03)
df_salary[['salary_year_avg', 'salary_year_inflated']]
```

```python
import ast

ast.literal_eval(df['job_skills'][1])
```

```python
def clean_list(skill_list):
    if pd.notna(skill_list):
        return ast.literal_eval(skill_list)

df['job_skills'] = df['job_skills'].apply(clean_list)
```


# Calculate projected salary next year

```python
df_salary['salary_year_inflated'] = df_salary['salary_year_avg'].apply(lambda salary: salary*1.03)

df_salary[['salary_year_avg', 'salary_year_inflated']]
```

```python
def projected_salary(row):
    if 'Senior' in row['job_title_short']:
        return 1.05*row['salary_year_avg']
    else:
        return 1.03*row['salary_year_avg']

df_salary['salary_year_inflated'] = df_salary.apply(projected_salary, axis=1)

df_salary[['job_title_short', 'salary_year_avg', 'salary_year_inflated']]

```

# combine with lambda

```python
df_salary['salary_year_inflated'] = df_salary.apply(lambda row: 1.05 * row['salary_year_avg'] if 'Senior' in row['job_title_short'] else 1.03 * row['salary_year_avg'], axis=1)

df_salary.head()
```

# Convert Data to String

```python
df['job_posted_date_str'] = df['job_posted_date'].apply(lambda x: x.strftime('%Y-%m-%d'))

df[['job_posted_date', 'job_posted_date_str']].head()
```

# Days Since Posted

```python
from datetime import datetime

now = datetime.now()
df['days_since_posted'] = df['job_posted_date'].apply(lambda date : (now - date).days)
df[['job_posted_date', 'days_since_posted']].head()
```

# Salary Category

```python
df_filtered = df.dropna(subset=['salary_year_avg']).copy()

df_filtered['salary_category'] = df['salary_year_avg'].apply(
    lambda x: 'Low' if x < 60000 
    else 'Medium' if x >= 60000 and x < 100000 
    else 'High')

df_filtered [['salary_year_avg','salary_category']]
```
